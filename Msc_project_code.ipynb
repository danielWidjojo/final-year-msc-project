```python

def avgW2vRepresentation( passage,length= 512 ):
    counterWordsVocab = 0
    w2vRepresentation = np.zeros(length)
    
    #print(w2vRepresentation)
    for words in passage:

        if words in modelW2v.key_to_index:
            counterWordsVocab +=1
            #print("trigger 1")
            #print(wv[words].shape)
            w2vRepresentation += modelW2v[words]
            #print(w2vRepresentation)
        #else:
        #    print("word not in vocab")
    if(counterWordsVocab != 0):
        #print("trigger")
        w2vRepresentation /= counterWordsVocab
    #print("middle",w2vRepresentation)
    return w2vRepresentation

def checkWord():
    wordsNotinList = []
    for searchTerm in df.columns:
        if searchTerm != 'date':
            words = preprocessing(searchTerm)
            for word in words:

                if word not in modelW2v.key_to_index:
                    wordsNotinList.append(word)
    return wordsNotinList

# a function to preprocess sentences
def preprocessing(sentences):
    ndivCorpusLower = sentences.lower()
    testQueries = str(ndivCorpusLower)
    # split to indiv word
#     splits = re.split(',|_|-|!| |\n|-|/', testQueries) 
    translator = testQueries.maketrans(string.punctuation, ' '*len(string.punctuation))
    testQueries = testQueries.translate(translator)
#     splits = re.split(',|_|-|!| |\n|-|/', testQueries) 
    splits = re.split(' |\n', testQueries) 
    # preprocess
    filteredSentence = []
    lemmatizer = WordNetLemmatizer()
    for w in splits:

        if w not in stopwords:
            if w!="" :

                w = lemmatizer.lemmatize(w)
                #print(type(w))
                filteredSentence.append(w)  #
    #print(filteredSentence)
    return filteredSentence
            
def similarity(word, positive, negative):
    # shld transfor by +1 and divide the entire thing by 2 to make sure there is no negative value
    preprocessed = preprocessing(word)
    wordToken = avgW2vRepresentation(preprocessed)
    positiveCosine = (modelW2v.cosine_similarities(wordToken, positive)+1)/2
    negativeCosine = (modelW2v.cosine_similarities(wordToken, negative)+1)/2
#     positiveCosine = (model.cosine_similarities(wordToken, positive))
#     negativeCosine = (model.cosine_similarities(wordToken, negative))
    gamma = 0.001
    similarityScore = np.sum(positiveCosine)/(np.sum(negativeCosine)+gamma)
    if (math.isnan(similarityScore) ):
        similarityScore= 0
    return similarityScore

def statisticResult (dicts):
    valuesSimilarity = np.array(list(dicts.values()))
    length = len(valuesSimilarity)
    mean = np.mean(valuesSimilarity)
    standardDev = np.std(valuesSimilarity)
    print(f"length of data {length}, mean {mean}, std dev {standardDev}")
    plt.hist(valuesSimilarity, bins=30)
    #sorting top result in descending order
    index = np.argsort(-1*valuesSimilarity)
    keySimilarity = np.array(list(dicts.keys()))
    sortedKey = keySimilarity[index]
    sortedVal = valuesSimilarity[index]
    dfStat = pd.DataFrame(sortedVal,columns = ['value'], index = sortedKey)
#     display(dfStat)
    return dfStat
    #.head(50)  
    
def pearsonCorrelationNoFilter(dfQuery, dfILIrates, dictSemantic):
    #get all the keys from semantically sorted queries
    sortedKeys = np.array(list(dictSemantic.keys()))
    
    dictPearson = {}
    counterNan = 0
    for keyword in sortedKeys:
        pearsonCorrelation = stats.pearsonr(dfQuery[keyword], dfILIrates['weight_ILI'])[0]
        
        if (math.isnan(pearsonCorrelation)):
            counterNan +=1
        else:
            dictPearson[keyword] = pearsonCorrelation
    print(counterNan)
    return dictPearson

def dateFilterDataFrame (dfGiven, startDate, endDate):
    # convert date column to date format
    dfGiven['date' ] = (pd.to_datetime(dfGiven['date' ])).dt.date
#     mask = (dfGiven['date' ] >= x) & (dfGiven['date' ] < x + datetime.timedelta(days=12))
    mask = (dfGiven['date' ] >= startDate) & (dfGiven['date' ] <= endDate)
    dfResult = dfGiven.loc[mask]
    return dfResult

def semanticFiltering (dictionaries, numbStd =0):
    #semantic filtering based on standard deviation
    valuesSimilarity = np.array(list(dictionaries.values()))
    keySimilarity = np.array(list(dictionaries.keys()))
    standardDev = np.std(valuesSimilarity)
    mean = np.mean(valuesSimilarity)
    index = valuesSimilarity > mean + (standardDev * numbStd)
    valuesSimilarity = valuesSimilarity[index]
    keySimilarity = keySimilarity[index]
    newDict = dict(zip(keySimilarity,valuesSimilarity))
    return newDict  

def pearsonCorrelation(dfQuery, dfILIrates, dictSemantic, rThreshold = 0):
    #get all the keys from semantically sorted queries
    sortedKeys = np.array(list(dictSemantic.keys()))
    print(len(sortedKeys))
    dictPearson = {}
    counterNan = 0
    for keyword in sortedKeys:
        pearsonCorrelation = stats.pearsonr(dfQuery[keyword], dfILIrates['weight_ILI'])[0]
#         print(keyword, pearsonCorrelation)
        if (math.isnan(pearsonCorrelation)):
            counterNan +=1
        if (pearsonCorrelation > rThreshold):
            dictPearson[keyword] = pearsonCorrelation
    print(counterNan)
    return dictPearson

#converting c to vector
def cTovector (cDict):
    positiveVect = np.zeros((len(c1['positive']),512))
    for index,contextWord in enumerate(c1['positive']):
        
        preprocessed = preprocessing(contextWord)
        positiveVect[index,:] = avgW2vRepresentation(preprocessed)
        
    negativeVect = np.zeros((len(c1['negative']),512))
    for index,contextWord in enumerate(c1['negative']):
        
        preprocessed = preprocessing(contextWord)
        negativeVect[index,:] = avgW2vRepresentation(preprocessed)
    return positiveVect,negativeVect
```

```python
def determineGaussianMixture(xpca,semanticFilterDict, thresholdGM = 0.002):
    numberToCluster = len(semanticFilterDict)
    import warnings
    for iteration in range (10):
        gm = GaussianMixture(n_components=numberToCluster, 
                                 random_state=123, 
                            n_init=5)
        with warnings.catch_warnings():
            warnings.filterwarnings('error')
            try:
                gm.fit(xpca)
                numberToClusterCurr = np.sum(np.round(gm.weights_, 3)>thresholdGM)

                if (numberToCluster == numberToClusterCurr):
                    break
                else:
                    numberToCluster = numberToClusterCurr
            except:
                numberToCluster -= int(0.25*numberToCluster)
    print(f"number of initial cluster {len(semanticFilterDict)} while final number of cluster is {numberToCluster}")
    return numberToCluster, gm
```

```python
def determineGaussianMixture2(xpca,semanticFilterDict,bic = True):
    bic_n, aic_n = bic_aic_score(xpca,maxVal = np.min((len(semanticFilterDict),100)), plot = False)
    if (bic):
        numberToCluster = bic_n
        print(f"bic is used cluster number {numberToCluster}")
    else:
        numberToCluster = aic_n
        print(f"aic is used cluster number {numberToCluster}")
    gm = GaussianMixture(n_components=numberToCluster, 
                                 random_state=123, 
                            n_init=5)
    return numberToCluster, gm
```

```python
def bic_aic_score(xpca, maxVal =100 , plot = True):
    n_range = np.arange(1,maxVal,2)
    # Create empty lists to store the BIC and AIC values
    bic_score = []
    aic_score = []

    # Loop through the range and fit a model
    for n in n_range:
        gm = GaussianMixture(n_components=n, 
                             random_state=123, 
                             n_init=1)
        gm.fit(xpca)

        # Append the BIC and AIC to the respective lists
        bic_score.append(gm.bic(xpca))
        aic_score.append(gm.aic(xpca))

    # Plot the BIC and AIC values together
    if (plot):
        fig, ax = plt.subplots(figsize=(12,8),nrows=1)
        ax.plot(n_range, bic_score, '-o', color='orange')
        ax.plot(n_range, aic_score, '-o', color='green')
        ax.set(xlabel='Number of Clusters', ylabel='Score')
        ax.set_xticks(n_range)
        ax.set_title('BIC and AIC Scores Per Number Of Clusters')
    print(f"number of cluster bic { n_range[np.argmin(bic_score)]} and aic {n_range[np.argmin(aic_score)]}")
    return n_range[np.argmin(bic_score)],n_range[np.argmin(aic_score)]
```

```python
def secondClustering(labelsCluster1,distMetric, threshold = 0.9):
    entireLabel = np.zeros(len(labelsCluster1))
    lastLabel = 0
    for cluster1 in np.unique(labelsCluster1):

        index = np.where(labelsCluster1 == cluster1)[0]
        if(len(index)>1):
    #         print(len(index))
            distRow = distMetric[index]
            distRowColumn = distRow[:,index]

            clusteringLevel2 = AgglomerativeClustering(affinity = 'precomputed', linkage='average',distance_threshold=threshold,n_clusters= None).fit((1-distRowColumn)) 
    #         print(np.unique(clusteringLevel2.labels_))
    #         print(np.unique(clusteringLevel2.labels_)+lastLabel)
            entireLabel[index] = clusteringLevel2.labels_+lastLabel
            lastLabel += len(np.unique(clusteringLevel2.labels_))

    #         print(lastLabel)
    #         print()
        else:
            entireLabel[index] = lastLabel

    #         print(lastLabel)
            lastLabel +=1
    #         print(lastLabel)
    return entireLabel
```

```python
def optimiseValidate3(train_dataloader,model,optimizer,criterion, validate=True,batch_number_eval=None):
    if (batch_number_eval == None):
        batch_number_eval = len(train_dataloader)
    lossPerEpoch = []
    lossDevPerEpoch = []
    evalCounter = 0
    for batch_index, dataloader in enumerate(train_dataloader):
        x,y,dates = dataloader
        monthConverted = np.array([int((datetime.date.fromtimestamp(date.int())+datetime.timedelta(days=1)).month) for date in dates])
        yearsConverted = np.array([int((datetime.date.fromtimestamp(date.int())+datetime.timedelta(days=1)).year) for date in dates])
        # selecting november until january to be the month to validate
        #(monthConverted>=12).any() or 
        
        if( (yearsConverted%2 == 0).any() and((monthConverted==12).any() or  (monthConverted<=4).any()) and validate):
            #set Evaluation mode to prevent training
            model.eval()
            #x = torch.flatten(x, start_dim=1)
            optimizer.zero_grad()
            # Forward pass
            evalCounter += 1
            with torch.no_grad():
                y_pred = model(x.to(device))
                lossDev = criterion(y_pred, y.to(device))
                lossDevPerEpoch.append(lossDev.cpu().detach())

            continue
        if(validate== False):
            lossDev = None
        model.train()
        #x = torch.flatten(x, start_dim=1)
        optimizer.zero_grad()
        # Forward pass
        y_pred = model(x.to(device))
        loss = criterion(y_pred, y.to(device))
        lossPerEpoch.append(loss.cpu().detach())
        # Backward pass
        loss.backward()
        optimizer.step()
#     print(f"eval {evalCounter} out of {batch_index+1}")
    return lossDevPerEpoch, lossPerEpoch
```

```python
# make parameter checked
class bestParameterClass():
    def __init__(self, folderLocation,name='bestParameter'):
        # Large value for best loss
        self.name = name
        self.location = folderLocation
        self.reset()
        
    def update(self, devLoss, parameter, model, parameterChecked=None,devArr = True):
        #for crossvalidating
        self.lastBestEpoch += 1
        #Check if parameter is the same or not
        
        if (parameter != self.parameter):
            #reset the counter when parameter changed
            self.parameter = parameter
            self.lastBestEpoch = 0
            self.bestLossPerParameter = 99999
        # per parameter 
        if (self.bestLossPerParameter > devLoss):
            self.lastBestEpoch = 0
            self.bestLossPerParameter = devLoss
        # for all parameter
        if (self.bestLoss > devLoss):
            self.bestParameterDict = parameter
            self.bestLoss = devLoss
            #add loss stuff to it
            parametersave = parameter.copy()
            parametersave.update(lossDict(devArr))
            self.bestParameterDict= parametersave
            savePickle(self.location, parametersave, self.name)
            torch.save(model,self.location+self.name)
            print()
            print(f"BestFNN {str(parameter)} with dev loss {self.bestLoss} at epoch {str(parametersave['epoch'])}")
        # for the checked parameter
        if (parameterChecked != None):
            if(self.parameterChecked != parameterChecked):
                
                self.parameterChecked = parameterChecked
                self.bestLossParameterChecked = 99999
            if (self.bestLossParameterChecked > devLoss):
                #dictionary to store the best per parameter
                parametersave = parameter.copy()
                parametersave.update(lossDict(devArr))
                self.bestParameterCheckedDict[str(parameterChecked)] = parametersave
                #updating the loss
                self.bestLossParameterChecked = devLoss
                #location to save
                locationPerParameter = self.location
                #make the folder to store the dictionary
                makeDir(locationPerParameter)
                #res = ','.join(key + str(val) for key, val in parameterChecked.items())
                res = 'bestPerParameter'
                savePickle(locationPerParameter, self.bestParameterCheckedDict, res)
                res = ','.join(key + str(val) for key, val in parameterChecked.items())
                torch.save(model,locationPerParameter+res)
        # printing the current best parameter
            
    def earlyBreakOff (self, threshold = 15):
        #print(self.lastBestEpoch)
        if(self.lastBestEpoch >threshold):
            value = True
        else:
            value = False
        return value
            
    def bestParameterReturn(self):
        return self.bestParameterDict,self.bestParameterCheckedDict
    def reset(self):
        self.bestLoss = 99999
        self.bestLossParameterChecked = 99999
        self.bestParameterDict = {}
        self.bestParameterCheckedDict = {}
        self.parameterChecked = None
        self.parameter = {}
        #Best loss per parameter
        self.bestLossPerParameter = 99999
        self.lastBestEpoch = 0
def parameterUpdate():
    parameter = {}
    parameter['learningrate'] = learning_rate 
    parameter['numberOfLayer'] = number_layer
    parameter['numberOfHiddenUnit'] =num_hidden_unit
    parameter['dropout'] = dropout
    parameter['numbQns'] = numbQns
    parameter['cluster'] = cluster
    parameter['cluster2'] = cluster2
    parameter['stdDev'] = stdDev
    parameter['threshold'] = threshold

    parameter['shuffleDataPoints'] = shuffleDataPoints
    # custom layer
    parameter['learnedWeights'] = learnedWeights
    # masked
    parameter['maskingStat'] = maskingStat
    return parameter
def lossDict(devArr):
    loss = {}
    loss['epoch'] = epoch
    if(devArr):
        loss['lossDev'] = devLossArray
    loss['lossArray'] = lossArray
    return loss

```

```python
class ClusterLayer(torch.nn.Module):
    def __init__(self,in_features,out_features,maskArr):
        """
        In the constructor we instantiate four parameters and assign them as
        member parameters.
        """
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.weights = torch.nn.Parameter(data=torch.Tensor(out_features,in_features ), requires_grad=True)
        #self.weights = torch.ones(out_features,in_features).double().to(device)
#         if bias:
#             self.bias = torch.nn.Parameter(torch.Tensor(out_features))
#         else:
#             self.register_parameter('bias', None)
        self.reset_parameters()
        self.maskArr = maskArr.double().to(device)
#         print(maskArr)
#         print(self.weights)
#         self.weights = self.weights*maskArr
    def reset_parameters(self):
        #torch.nn.init.kaiming_uniform_(self.weights, a=math.sqrt(5))
        # a only for leaky relu
        #, a=math.sqrt(5)
        torch.nn.init.uniform_(self.weights, 0, 1)
        
#         if self.bias is not None:
#             fan_in, _torch.nn.init._calculate_fan_in_and_fan_out(self.weights)
#             bound = 1 / math.sqrt(fan_in)
#             torch.nn.init.uniform_(self.bias, -bound, bound)
        
    def forward(self, x):
        #print(self.weights)
#         if(mask!=False):

        weights = (self.weights*self.maskArr)
        
        #weights = self.weights*maskArr
        output = x @ weights.t()
        return output
```

```python
# prob the third model that integrate both method
class Feedforward3(torch.nn.Module):
        def __init__(self, input_size, hidden_size,dropout = 0,layerNumb = 1,maskArr=None):
            super(Feedforward3, self).__init__()
            # size of input layer is two dimension
            # the first is typically 28 corresponding to the training window
            # the second one is typically the number of feature
            self.input_size = input_size
            self.hidden_size  = hidden_size
            self.maskArr = maskArr
            
            if (maskArr is not None):
#                 print("mask is used")
                self.cluster_size = maskArr.shape[0]
                self.maskArr = torch.Tensor(maskArr)
                
                self.clusterLayer = ClusterLayer(self.input_size[1], self.cluster_size,self.maskArr)
                #self.dense1_bn = torch.nn.BatchNorm1d(input_size[0])
                self.fc0 = torch.nn.Linear(self.cluster_size*input_size[0], self.hidden_size)
            else:
                
                self.fc0 = torch.nn.Linear(input_size[1]*input_size[0], self.hidden_size)
            self.layerNumb = layerNumb
            
            
            
            #self.relu = torch.nn.ReLU()
            for layer in range(layerNumb-1):
#                 print(layer)
                
                setattr(self, f"fc{layer+1}", torch.nn.Linear(self.hidden_size, self.hidden_size))

            setattr(self, f"fc{layerNumb}", torch.nn.Linear(self.hidden_size, 1))
            self.droput = torch.nn.Dropout(dropout)
            
            #self.norm = torch.nn.BatchNorm1d(self.hidden_size)
            #self.sigmoid = torch.nn.Sigmoid()
        
        def forward(self, x):
            #activation = F.relu
            if(self.maskArr is not None):
                output = self.clusterLayer(x)
                #output = self.dense1_bn(output)
                output = torch.flatten(output, start_dim=1)
                output = F.leaky_relu(output)
                #flatten the output
                
            else:
                output = x
                output = torch.flatten(output, start_dim=1)
            #flatten the output
            
            output = torch.flatten(output, start_dim=1)
            output = F.leaky_relu(self.fc0(output))
            output = self.droput(output)
            for layer in range(self.layerNumb-1):
                output = F.leaky_relu(getattr(self, f"fc{layer+1}")(output))
                output = self.droput(output)
            output = getattr(self, f"fc{self.layerNumb}")(output)

            return output
```

```python
def updatingStatisticalResult(dfsummary,dfOverallData):
    dfsummaryMean = dfOverallData.mean()
    dfsummaryMean['type'] = 'mean'
    dfsummaryMax = dfOverallData.max()
    dfsummaryMax['type'] = 'max'
    dfsummaryMin = dfOverallData.min()
    dfsummaryMin['type'] = 'min'
    dfsummaryStd = dfOverallData.std()
    dfsummaryStd['type'] = 'std'
    dfsummaryMean['start year'] =dfsummaryMax['start year'] 
    dfsummaryMean['end year'] =dfsummaryMax ['end year'] 
    dfsummaryStd['start year'] =dfsummaryMax ['start year'] 
    dfsummaryStd['end year'] =dfsummaryMax ['end year'] 
    dfsummary = dfsummary.append(dfsummaryMean,ignore_index=True)
    dfsummary = dfsummary.append(dfsummaryStd,ignore_index=True)
    dfsummary = dfsummary.append(dfsummaryMax,ignore_index=True)
    dfsummary = dfsummary.append(dfsummaryMin,ignore_index=True)
    return dfsummary
    
```

```python
def printStatus(bestOverallDict):
    num_hidden_unit = bestOverallDict['numberOfHiddenUnit']
    dropout = bestOverallDict['dropout']
    number_layer = bestOverallDict['numberOfLayer']
    learning_rate = bestOverallDict['learningrate']
    epochs = bestOverallDict['epoch']
    devLossArray = bestOverallDict['lossDev']
    lossArray = bestOverallDict['lossArray']
    clusterSizes = bestOverallDict['cluster']
    stdDev = bestOverallDict['stdDev']
    learnedWeights = bestOverallDict['learnedWeights']
    threshold = bestOverallDict['threshold']
    print(f"Best FNN learningrate {learning_rate},numberOfLayer {number_layer},numberOfHiddenUnit {num_hidden_unit},dropout {dropout}, epocs {epochs}")
    print(f"Best cluster {clusterSizes} at std Dev {stdDev}")
    print(f"Best test loss {devLossArray[epochs]} at train loss {lossArray[epochs]}")
    print(f"Best train loss {min(lossArray)}")
    print(f"Total training epoch {len(lossArray)}")
    print(f"learned weights {learnedWeights}")
    print(f"threshold {threshold}")
    return devLossArray[epochs]
```

```python
# Generate line of data per try to be appended
def lineOfData(dfPlot):
    dfIter = presentData(dfPlot)
    arrayInput = [dfPlot.iloc[0,0].year,dfPlot.iloc[-2,0].year]
    arrayInput.extend(dfIter.iloc[-1,1:])
    dfTest = pd.DataFrame([arrayInput], columns = ['start year', 'end year','pearson correlation','MAE',"MAPE","MSE"])
    return dfTest
```
